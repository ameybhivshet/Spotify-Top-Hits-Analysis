## B) Finding song trends over the years

```python
#key,speechiness,mode,tempo, duration_ms,time_signature,loudness are dropped from caclulation
avg_yearwise = data.drop(['key','mode','speechiness','tempo','duration_ms','time_signature','loudness'],axis=1, inplace=False).groupby('playlist', as_index=False).mean()
```

```python
#Splitting top_playlist to get the year column
avg_yearwise[['top','Year']] = avg_yearwise.playlist.str.split("_",expand=True) 
avg_yearwise.drop(['top','playlist'], axis= 1, inplace= True)
avg_yearwise.head()
```

```python
#Melting the dataframe so that plotting becomes easier
melted_data = avg_yearwise.melt(id_vars='Year')
melted_data.head()
```

```python
plt.figure(figsize=(15,6))
plt.title("Song Trends Over Time", fontsize=15)

sns.lineplot(x='Year', y='value', hue='variable', data=melted_data)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize =14 )
```

The average energy levels have been dropping gradually from 2010 while we can see a increase in the average danceability values.

There was a sudden increase in the mean acousticness form 2017. 


The line chart confirms that danceability and energy are important for a song to make it to the top charts and there has to be some vocal content as well.
